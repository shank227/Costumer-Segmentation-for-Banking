```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Shashank\Banking_Segment\Data\Bank_dataset\bank-additional-full.csv", sep=";")
```

```python
#missing value analysis
(df == "unknown").sum()

unknown_counts = (df == "unknown").sum()
unknown_counts[unknown_counts > 0]
```

```python
#feature classification
categorical_cols = df.select_dtypes(include='object').columns
numerical_cols = df.select_dtypes(include=['int64','float64']).columns

print("Categorical Features:")
print(categorical_cols)

print("\nNumerical Features:")
print(numerical_cols)
```

```python
#Unique Value Analysis
for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].nunique())
```

```python
for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].unique())
```

```python
#Removing Target variable (the subscription column)
df_ml = df.drop('y', axis=1)
df_ml.shape
```

```python
categorical_cols = df_ml.select_dtypes(include='object').columns
numerical_cols = df_ml.select_dtypes(include=['int64','float64']).columns

print(categorical_cols)
print("\n")
print(numerical_cols)
```

```python
#Data Encoding (1-hot encoding)
df_encoded = pd.get_dummies(
    df_ml,
    columns=categorical_cols,
    drop_first=True
)

df_encoded.shape
df_encoded.head()
```

```python
df_encoded.columns
```

```python
print("Original Shape:", df_ml.shape)
print("Encoded Shape:", df_encoded.shape)
```

```python
df_encoded.select_dtypes(include='object')
```

```python
df_encoded.to_csv("encoded_dataset.csv", index=False)
```

```python
#Feature Engineering (age groups)
df_encoded['age_group'] = pd.cut(
    df['age'],
    bins=[0,25,40,60,100],
    labels=[0,1,2,3]
)

df_encoded['age_group'].value_counts()
```

```python
#Feature Engineering (contacted before)

df_encoded['contacted_before'] = (
    df['previous'] > 0
).astype(int)

df_encoded['contacted_before'].value_counts()
```

```python
median_campaign = df['campaign'].median()

df_encoded['high_campaign'] = (
    df['campaign'] > median_campaign
).astype(int)

df_encoded['high_campaign'].value_counts()
```
