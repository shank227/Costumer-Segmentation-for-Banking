```python
#Removing Target variable (the subscription column)
df_ml = df.drop('y', axis=1)
df_ml.shape
```

```python
categorical_cols = df_ml.select_dtypes(include='object').columns
numerical_cols = df_ml.select_dtypes(include=['int64','float64']).columns

print(categorical_cols)
print("\n")
print(numerical_cols)
```

```python
#Data Encoding (1-hot encoding)
df_encoded = pd.get_dummies(
    df_ml,
    columns=categorical_cols,
    drop_first=True
)

df_encoded.shape
df_encoded.head()
```

```python
df_encoded.columns
```

```python
print("Original Shape:", df_ml.shape)
print("Encoded Shape:", df_encoded.shape)
```

```python
df_encoded.select_dtypes(include='object')
```

```python
df_encoded.to_csv("encoded_dataset.csv", index=False)
```

```python
#Feature Engineering (age groups)
df_encoded['age_group'] = pd.cut(
    df['age'],
    bins=[0,25,40,60,100],
    labels=[0,1,2,3]
)

df_encoded['age_group'].value_counts()
```

```python
#Age Labeling
age_group_labels = {
    0: "Young",
    1: "Adult",
    2: "Middle_Age",
    3: "Senior"
}

df_encoded['age_group_label'] = (
    df_encoded['age_group']
    .map(age_group_labels)
)

df_encoded['age_group_label'].value_counts()
```

```python
#Feature Engineering (contacted before)

df_encoded['contacted_before'] = (
    df['previous'] > 0
).astype(int)

df_encoded['contacted_before'].value_counts()
```

```python
#contacted_before labeling
contact_labels = {
    0: "No",
    1: "Yes"
}

df_encoded['contacted_before_label'] = (
    df_encoded['contacted_before']
    .map(contact_labels)
)

df_encoded['contacted_before_label'].value_counts()
```

```python
median_campaign = df['campaign'].median()

df_encoded['high_campaign'] = (
    df['campaign'] > median_campaign
).astype(int)

df_encoded['high_campaign'].value_counts()
```

```python
campaign_labels = {
    0: "Low_Campaign",
    1: "High_Campaign"
}

df_encoded['high_campaign_label'] = (
    df_encoded['high_campaign']
    .map(campaign_labels)
)

df_encoded['high_campaign_label'].value_counts()
```

```python
df_encoded.info()
```

```python
#Converting Boolean columns to integers

bool_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[bool_cols] = df_encoded[bool_cols].astype(int)

df_encoded.info()
```

```python
#Scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded)
```

```python
scaled_df = pd.DataFrame(
    scaled_data,
    columns=df_encoded.columns
)
```
