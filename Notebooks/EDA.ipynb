```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Shashank\Banking_Segment\Data\Bank_dataset\bank-additional-full.csv", sep=";")
```

```python
df.head
```

```python
df.info()
```

```python
plt.hist(df["age"], bins = 20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
median_age = df['age'].median()
print(f"Median Age: {median_age}")
```

```python
df["job"].value_counts().plot(kind="bar")
plt.title("Costumer Distribution by Job")
plt.show()
```

```python
df["education"].value_counts().plot(kind="bar")
plt.title("Educational Background of Customers")
plt.show()
```

```python
df["housing"].value_counts().plot(kind="bar")
plt.title("Housing Status")
plt.show()
```

```python
df["y"].value_counts().plot(kind="bar")
plt.title("Term Subscription")
plt.show()
```

```python
pd.crosstab(df["job"], df["y"]).plot(kind="bar")
plt.title("Job : Subscription Ratio")
plt.show()
```
