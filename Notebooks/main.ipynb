```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"Data\Bank dataset\bank-additional-full.csv", sep=';')
```

```python
df.head
```

```python
df.info()
```

```python
plt.hist(df["age"], bins = 20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df["job"].value_counts().plot(kind="bar")
plt.title("Costumer Distribution by Job")
plt.show()
```

```python
df["education"].value_counts().plot(kind="bar")
plt.title("Education Chart")
plt.show()
```

```python
df["housing"].value_counts().plot(kind="bar")
plt.title("housing status")
plt.show()
```

```python
df["y"].value_counts().plot(kind="bar")
plt.title("Subscription")
plt.show()
```

```python
pd.crosstab(df["job"], df["y"]).plot(kind="bar")
plt.show()
```
